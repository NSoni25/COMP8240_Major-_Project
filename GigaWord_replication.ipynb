Import Autotokenizer and AutomodelforSeq2SeqLM library for model generation.Autotokenizer is used to generate tokens of the sentences and AutomodelforSeq2SeqLM is used for pre-trained model generation

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import torch
from rouge_score import rouge_scorer
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
import re
```

```python
##Pegasus large model used in this approach
tokenizer = AutoTokenizer.from_pretrained("google/pegasus-large")

model = AutoModelForSeq2SeqLM.from_pretrained("google/pegasus-large")
```

```python
!pip install datasets
```

```python
#gigaword dataset doenloaded
from datasets import load_dataset

dataset = load_dataset("gigaword")
```

```python
print(dataset)
```

```python
import pandas as pd
```

```python
gigaword_list = ['zairean rebels , led by laurent-desire kabila , on saturday rejected calls by the united nations for a ceasefire , saying it could only be called after talks with kinshasa .',
                 'a zairean diplomat was among three african men arrested for smuggling ## kilograms -lrb- ### pounds -rrb- of marijuana into sweden , the swedish news agency tt reported friday .',
                 'prime minister benjamin netanyahu has proposed completing talks on the final status of palestinian territories in the next six months instead of the scheduled two years , according to israeli public television .']
```

```python
document = []
summary = []
df = pd.DataFrame(columns = ['document','summary'])
for i in range(0,len(dataset['test']['document'])):
  for j in gigaword_list:
    if j in dataset['test']['document'][i]:
      summary.append(dataset['test']['summary'][i])
      document.append(dataset['test']['document'][i])
df['document'] = document
df['summary'] = summary
df     
```

```python
#Score for document id 269
inputs = tokenizer(document[0],truncation=True, padding="longest", return_tensors='pt')
prediction = model.generate(**inputs)
prediction = tokenizer.batch_decode(prediction)
scorer = rouge_scorer.RougeScorer(['rouge1', 'rougeL'], use_stemmer=True)
scores = scorer.score(str(summary[0]),str(prediction))
print("Document",document[0] )
print('Dataset summary: ',summary[0])
print('Model Prediction: ',prediction)
print('Fmeasure as F1 scores: ',scores['rouge1'].fmeasure)
```

```python
#Score for document ID 198
inputs = tokenizer(document[1],truncation=True, padding="longest", return_tensors='pt')
prediction = model.generate(**inputs)
prediction = tokenizer.batch_decode(prediction)
scorer = rouge_scorer.RougeScorer(['rouge1', 'rougeL'], use_stemmer=True)
scores = scorer.score(str(summary[1]),str(prediction))
print("Document",document[1] )
print('Dataset summary: ',summary[1])
print('Model Prediction: ',prediction)
print('Fmeasure as F1 scores: ',scores['rouge1'].fmeasure)
```
